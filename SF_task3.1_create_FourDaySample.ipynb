```python
for month in np.unique(my_data.mo):
    print month
    if month == 4:
        apr_data = my_data[(my_data.mo == 4) & (my_data.day == 24)]
        apr_data['date_id'] = 1
        print len(apr_data)
        print np.unique(apr_data.mo)
    if month == 5:
        may_data = my_data[(my_data.mo == 5) & (my_data.day == 2)]
        may_data['date_id'] = 2
        print len(may_data)
        print np.unique(may_data.mo)
    if month == 9:
        sep_data = my_data[(my_data.mo == 9) & (my_data.day == 20)]
        sep_data['date_id'] = 3
        print len(sep_data)
        print np.unique(sep_data.mo)
    if month == 10:
        oct_data = my_data[(my_data.mo == 10) & (my_data.day == 10)]
        oct_data['date_id'] = 4
        print len(oct_data)
        print np.unique(oct_data.mo) 
```
